```
x = 1
while x < 10:
  print(x)
  #x = x+1
  x += 1
```

```
x = 1
while x < 10:
  print(x)
  if x == 5:
    break
  #x = x+1
  x += 1
```

```
x = 1
while x < 10:
  x += 1
  if x == 5:
    continue
  print(x)
```
