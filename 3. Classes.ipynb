```
class Car():

  # constructor
  def __init__(self, color, brand, engine_type):
    self.color = color
    self.brand = brand
    self.engine_type = engine_type
  
  def change_color(self, new_color):
    self.color = new_color
```

```
mycar = Car(color="White", brand="Toyota", engine_type="Diesel")
```

```
mycar.color
```

```
class Father():

  def __init__(self):
    print("This is father constructor")

  def call(self):
    print("I am father")

class Mother():
  def calling(self):
    print("I am mother")

class Son(Father, Mother):

  def __init__(self):
    print("This is son constructor")

  def call_me(self):
    print("This is Rahul")
```

```
s = Son()
```

```
s.calling()
```
