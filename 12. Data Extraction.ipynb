### Working with XML data

```
import xml.etree.cElementTree as et
```

```
xml_data = et.parse('NCHS_-_Death_rates_and_life_expectancy_at_birth.xml')
```

```
xml_data
```

```
root_node = xml_data.getroot()
```

```
root_node
```

```
root_node.getchildren()
```

```
root_node.getchildren()[0]
```

```
root_node.getchildren()[0].getchildren()[0].find('year').text
```

```
year = []
race = []
gender = []
ale = []
mortality = []
for node in root_node.getchildren()[0].getchildren():
  year.append(node.find('year').text)
  race.append(node.find('race').text)
  gender.append(node.find('sex').text)
  try:
    ale.append(node.find('average_life_expectancy').text)
  except:
    ale.append(None)
  mortality.append(node.find('mortality').text)
  
```

```
df = pd.DataFrame({'Year':year, 'Race':race, 'Gender':gender, 'ALE':ale, 'mortality':mortality})
```

```
df.head()
```

```
df.dtypes
```

### Working with text files

```
file = open('wx-data.txt', 'r')
```

```
line = file.readline()
```

```
line
```

```
import re
```

```
value = re.search('(Dm=)[0-9]+', line).group()
```

```
Dm = value.replace('Dm=','')
```

```
Dm
```

```
Dm = []
for line in file:
  try:
    value = re.search('(Dm=)[0-9]+', line).group()
    temp = value.replace('Dm=','')
    Dm.append(int(temp))
  except Exception as e:
    Dm.append(None)
  
```
