### add x and y labels

```
plt.scatter(x=data['rm'], y=data['medv'])
plt.title('rm vs medv')
plt.xlabel('rm')
plt.ylabel('medv')
plt.show()
```

```
plt.scatter(x=data['rm'], y=data['medv'], c=data['rad'])
plt.title('rm vs medv')
plt.xlabel('rm')
plt.ylabel('medv')
plt.show()
```

```
plt.scatter(x=data['rm'], y=data['medv'], c=data['rad'], s=data['age'])
plt.title('rm vs medv')
plt.xlabel('rm')
plt.ylabel('medv')
plt.show()
```

```
fig, ax = plt.subplots()
plot = ax.scatter(x=data['rm'], y=data['medv'], c=data['rad'], s=data['age'])
handles, labels = plot.legend_elements(prop='sizes')
legend1 = ax.legend(handles, labels, loc='upper right', title='Sizes')
plt.title('rm vs medv')
plt.xlabel('rm')
plt.ylabel('medv')
plt.show()
```

```
fig, ax = plt.subplots()
plot = ax.scatter(x=data['rm'], y=data['medv'], c=data['rad'], s=data['age'])


legend2 = ax.legend(*plot.legend_elements(), loc='lower left', title='Classes')
ax.add_artist(legend2)

handles, labels = plot.legend_elements(prop='sizes')
legend1 = ax.legend(handles, labels, loc='upper right', title='Sizes')

plt.title('rm vs medv')
plt.xlabel('rm')
plt.ylabel('medv')
plt.show()
```

```
fig, ax = plt.subplots(figsize=(15,10))
plot = ax.scatter(x=data['rm'], y=data['medv'], c=data['rad'], s=data['age'])


legend2 = ax.legend(*plot.legend_elements(), loc='lower left', title='Classes')
ax.add_artist(legend2)

handles, labels = plot.legend_elements(prop='sizes')
legend1 = ax.legend(handles, labels, loc='upper right', title='Sizes')

plt.title('rm vs medv')
plt.xlabel('rm')
plt.ylabel('medv')
plt.show()
```

```
simple_data = pd.DataFrame({'Markets':['US', 'UK', 'IN', 'AU'], 'Sales':[3456,3345,2356,4567], 'Targets':[3567, 3456, 2456, 4666]})
```

```
plt.bar(x=simple_data['Markets'], height=simple_data['Sales'])
```

```
plt.bar(x=simple_data['Markets'], height=simple_data['Sales'])
plt.plot(simple_data['Targets'])
plt.show()
```

```
plt.bar(x=simple_data['Markets'], height=simple_data['Sales'])
plt.scatter(x=simple_data['Markets'], y=simple_data['Targets'])
plt.plot(simple_data['Targets'])
plt.show()
```

```
fig, (ax1, ax2) = plt.subplots(2)
ax1.scatter(x=simple_data['Markets'], y=simple_data['Targets'])
ax2.bar(x=simple_data['Markets'], height=simple_data['Sales'])
```
