```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp
```

```python
# import data
data = pd.read_excel('Raw Data_28.02.2021.xlsx')
```

```python
data.head()
```

```python
def one_sample_one_tailed(sample_data, popmean, alpha=0.05, alternative='greater'):
    t, p = ttest_1samp(sample_data, popmean)
    #print ('t:',t)
    #print ('p:',p)
    if alternative == 'greater' and (p/2 < alpha):
        #print ('Reject Null Hypothesis for greater-than test')
        return np.mean(sample_data),t,p/2,'reject'
    else:
        #print ('Accept Null Hypothesis for greater-than test')
        return np.mean(sample_data),t,p/2,'accept'
```

```python
sample_mean_list, tvalue_list, pvalue_list, hyp_list, factor_list = [], [], [], [], []
```

```python
for i in range(3,len(data.columns)):
    df = data.iloc[:,i]
    factor = df.name
    sample_mean,tvalue, pvalue, hypothesis = one_sample_one_tailed(df,3)
    sample_mean_list.append(sample_mean)
    tvalue_list.append(tvalue)
    pvalue_list.append(pvalue)
    hyp_list.append(hypothesis)
    factor_list.append(factor)
```

```python
final_df = pd.DataFrame({'Factor':factor_list,'Sample Mean':sample_mean_list, 'tvalue':tvalue_list, 'pvalue':pvalue_list, 'hypothesis':hyp_list})
```

```python
final_df.to_csv('ttest for factors report.csv', index=False)
```
