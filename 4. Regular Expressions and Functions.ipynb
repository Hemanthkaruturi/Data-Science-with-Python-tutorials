```
def smart_divisor(divison):
  def inner(x,y):
    if y == 0:
      print("division is not possible")
    else:
      return divison(x,y)
  return inner

def divison(x,y):
  return x/y
```

```
divison = smart_divisor(divison)
```

```
divison
```

```
divison(4,2)
```

```
divison(4,0)
```

```
def smart_divisor(divison):
  def inner(x,y):
    if y == 0:
      print("division is not possible")
    else:
      return divison(x,y)
  return inner

@smart_divisor
def divison(x,y):
  return x/y
```

```
divison(4,2)
```

```
divison(4,0)
```
